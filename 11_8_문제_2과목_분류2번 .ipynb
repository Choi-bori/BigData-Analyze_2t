# 문제2

## 흡연 상태 예측
- 건강 지표에 대한 정보를 바탕으로 환자의 흡연 상태를 예측

- 평가지표: roc-auc

- smoking : 흡연 - 1 / 비흡연 - 0


문제 : id랑 smoking : 1, 0인 2개의 컬럼으로 제출

```python
# 문제에서 주어지는 부분
import pandas as pd
train = pd.read_csv('11.08_연습문제_2/train.csv') ## 파일 경로 바꾸기
test = pd.read_csv('11.08_연습문제_2/test.csv') ## 파일 경로 바꾸기
```

```python
df = train
df
```

```python
X = df.drop(columns = 'smoking')
Y = df['smoking']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size = 0.7)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = DecisionTreeClassifier()

model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pred, y_test))
```

```python

test = pd.read_csv('11.08_연습문제_2/test.csv') ## 파일 경로 바꾸기
test['smoking'] = model.predict(test)

# test[['id','smoking'].to_csv('1108_2classification.csv')
```
