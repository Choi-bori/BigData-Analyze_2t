```python
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
test
```

```python
df1 = train
df1.columns
```

```python
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
le = LabelEncoder()
```

```python
df1['Sex'] = le.fit_transform(df1['Sex'])
test['Sex'] = le.fit_transform(test['Sex'])
```

```python
sca = MinMaxScaler()
```

```python
df1[['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight','Viscera Weight', 'Shell Weight']] = sca.fit_transform(df1[['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight','Viscera Weight', 'Shell Weight']])
test[['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight','Viscera Weight', 'Shell Weight']] = sca.fit_transform(test[['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight','Viscera Weight', 'Shell Weight']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1[['Sex', 'Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight','Viscera Weight', 'Shell Weight']]
Y = df1['Age']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python

pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,pred)
mae
```

```python
result = model.predict(test.drop(columns = 'id'))
submit = pd.DataFrame({'id':test['id'], 'Age' : result})
submit
```

```python
#기본 정보 조회
train.info()
test.info()
```

```python
# id를 test에서 pop해서 저장
# train도 동일하게 id를 제거, Sex가 유일하게 object변수여서 전처리 필요하지만 나이에 나이는 무방하다고 판단후 제거하고 예측(시험볼때 가장 쉽게 풀기위해서)
id = test.pop('id')
y = train.pop('Age')

x= train.drop(columns = ['id','Sex'])
test = test.drop(columns = ['Sex'])
```

```python
# min-max 진행할때 코드 추가
# 실제 시험에서는 제거해도 무방할듯함(미세한 차이기 때문에 40점 받을거 같음)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(x)
test_scaled = scaler.transform(test)  # 같은 스케일링을 테스트 데이터에 적용
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_val,y_tr,y_val = train_test_split(x,y,random_state = 42 , test_size = 0.2)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 42,n_estimators = 500 , max_depth = 12)

rf.fit(x_tr,y_tr)
pred = rf.predict(x_val)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_val,pred)
mae
```

```python
# 결과제출
result = rf.predict(test)
result
submit = pd.DataFrame({'id':id, 'yield' : result})
submit.to_csv('submit_mm_scaler.csv' , index = False)
```
